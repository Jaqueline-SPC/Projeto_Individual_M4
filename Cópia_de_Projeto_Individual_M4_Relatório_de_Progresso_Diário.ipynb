#Relatórios de outros itens

```python
import matplotlib.pyplot as plt

# Dados fornecidos na tabela
dados = {
    'Dia': ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado', 'Domingo'],
    'Horas Trabalhadas': [6, 7, 8, 6, 7, 5, 4],
    'Bugs Corrigidos': [3, 2, 1, 4, 3, 2, 1],
    'Tarefas Concluídas': [5, 4, 6, 4, 5, 3, 2]
}

# Criando o DataFrame com os dados
df_programacao = pd.DataFrame(dados)

# Calculando as métricas
total_horas_trabalhadas = df_programacao['Horas Trabalhadas'].sum()
media_diaria_horas_trabalhadas = df_programacao['Horas Trabalhadas'].mean()

total_bugs_corrigidos = df_programacao['Bugs Corrigidos'].sum()
media_diaria_bugs_corrigidos = df_programacao['Bugs Corrigidos'].mean()

total_tarefas_concluidas = df_programacao['Tarefas Concluídas'].sum()
media_diaria_tarefas_concluidas = df_programacao['Tarefas Concluídas'].mean()

# Calculando a Produtividade Diária (Tarefas Concluídas por Hora)
df_programacao['Produtividade Diária'] = df_programacao['Tarefas Concluídas'] / df_programacao['Horas Trabalhadas']

# Gráficos
fig, axs = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Relatório de Progresso Diário')

# Gráfico de Barras - Total de Horas Trabalhadas e Média Diária de Horas Trabalhadas
axs[0, 0].bar(df_programacao['Dia'], df_programacao['Horas Trabalhadas'], color='DarkOrange', label='Horas Trabalhadas')
axs[0, 0].axhline(y=media_diaria_horas_trabalhadas, color='red', linestyle='--', label='Média Diária')
axs[0, 0].set_title('Total de Horas Trabalhadas e Média Diária de Horas Trabalhadas')
axs[0, 0].set_xlabel('Dias da semana')
axs[0, 0].set_ylabel('Horas Trabalhadas')
axs[0, 0].legend()

# Gráfico de Barras - Total de Bugs Corrigidos e Média Diária de Bugs Corrigidos
axs[0, 1].bar(df_programacao['Dia'], df_programacao['Bugs Corrigidos'], color='DodgerBlue', label='Bugs Corrigidos')
axs[0, 1].axhline(y=media_diaria_bugs_corrigidos, color='red', linestyle='--', label='Média Diária')
axs[0, 1].set_title('Total de Bugs Corrigidos e Média Diária de Bugs Corrigidos')
axs[0, 1].set_xlabel('Dias da semana')
axs[0, 1].set_ylabel('Bugs Corrigidos')
axs[0, 1].legend()

# Gráfico de Barras - Total de Tarefas Concluídas e Média Diária de Tarefas Concluídas
axs[1, 0].bar(df_programacao['Dia'], df_programacao['Tarefas Concluídas'], color='PaleVioletRed', label='Tarefas Concluídas')
axs[1, 0].axhline(y=media_diaria_tarefas_concluidas, color='red', linestyle='--', label='Média Diária')
axs[1, 0].set_title('Total de Tarefas Concluídas e Média Diária de Tarefas Concluídas')
axs[1, 0].set_xlabel('Dias da semana')
axs[1, 0].set_ylabel('Tarefas Concluídas')
axs[1, 0].legend()

# Gráfico de Linha - Produtividade Diária
axs[1, 1].plot(df_programacao['Dia'], df_programacao['Produtividade Diária'], marker='o', color='green', label='Produtividade Diária')
axs[1, 1].set_title('Produtividade Diária (Tarefas Concluídas por Hora)')
axs[1, 1].set_xlabel('Dias da semana')
axs[1, 1].set_ylabel('Produtividade Diária')
axs[1, 1].legend()

# Ajustando espaçamento entre subplots
plt.tight_layout()

# Exibindo o resultado do gráfico
plt.show()
```
